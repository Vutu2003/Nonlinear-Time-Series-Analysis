```python
import sys
import os
import matplotlib.pyplot as plt

# Đảm bảo Jupyter tìm thấy thư mục core_ntsa
sys.path.append(os.path.abspath(".."))
from core_ntsa.generators import simulate_lorenz, simulate_rossler, simulate_henon

# 1. Sinh dữ liệu từ 3 hệ thống
print("Đang mô phỏng hệ Lorenz...")
lorenz_data = simulate_lorenz(t_span=100.0, dt=0.01)

print("Đang mô phỏng hệ Rössler...")
rossler_data = simulate_rossler(t_span=500.0, dt=0.1)

print("Đang mô phỏng hệ Hénon...")
henon_data = simulate_henon(n_iterations=10000)

print("Đã sinh xong dữ liệu. Bắt đầu vẽ đồ thị...")

# 2. Khởi tạo không gian vẽ (1 hàng, 3 cột)
fig = plt.figure(figsize=(12, 6))

# --- ĐỒ THỊ 1: LORENZ (3D) ---
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot(lorenz_data[0], lorenz_data[1], lorenz_data[2], lw=0.5, color='darkblue')
ax1.set_title("Lorenz Attractor\n(Continuous, 3D)", fontsize=14, fontweight='bold')
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_zlabel("Z")
# 

# --- ĐỒ THỊ 2: RÖSSLER (3D) ---
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot(rossler_data[0], rossler_data[1], rossler_data[2], lw=0.5, color='darkred')
ax2.set_title("Rössler Attractor\n(Continuous, 3D)", fontsize=14, fontweight='bold')
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_zlabel("Z")
# 

# --- ĐỒ THỊ 3: HÉNON (2D) ---
# Chú ý: Dùng dấu chấm ('.') hoặc scatter vì nó là dữ liệu rời rạc
ax3 = fig.add_subplot(133)
ax3.plot(henon_data[0], henon_data[1], '.', markersize=1, color='darkgreen')
ax3.set_title("Hénon Map\n(Discrete, 2D)", fontsize=14, fontweight='bold')
ax3.set_xlabel("X")
ax3.set_ylabel("Y")
ax3.grid(True, linestyle='--', alpha=0.5)
# 

# Tối ưu hóa khoảng cách giữa các đồ thị
plt.tight_layout()
plt.show()
```
